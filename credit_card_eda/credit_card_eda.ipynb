# 1. Обзор данных

```python
# создадим функцию, которая будет 
# печатать базовую информацию
# о каждой таблице

path = '/mnt/c/Users/koles/Downloads/datasets/credit_cards_eda/'

def first_view():
    files = os.listdir(path)
    dfs = []
    for file in files:
        df = pd.read_csv(path+file)
        dfs.append(df)
        
        print(file, '\n')
        display(df.head())
        display(df.info())
        display(df.describe(include='all'))
        print('Количество пропущеных значений:', df.isna().sum(), sep='\n')
        print(f'\nКоличество дубликатов - {df.duplicated().sum()}')
        print('\n\n')
        
    return dfs

files = first_view()

# и сохраним каждую таблицу в
# python переменные
cust_ac, repayment, spend = files
```

# 2. Предобработка данных

### Customer Acquisition

```python
# изменим названия колонок
cust_ac.columns = [i.lower() for i in cust_ac.columns]

# удалим колонку с названием `no`
cust_ac.drop('no', axis=1, inplace=True)

# проверим результат
cust_ac.head()
```

```python
# проанализируем владельцев карточек,
# чей возраст меньше 18 лет

under_18 = cust_ac.query('age < 18')

display(under_18, under_18.shape)
```

Всего таких пользователей 6. Трое имеют карточку со стратусом `Gold`, двое - `Platinum`, один - `Silver`. 4 из 6 имеют максимальны кредитный лимит 500 тыс. Все шестеро самозаннятые или работают на приватную либо государственную компанию, что невозможно для людей младше 18 лет.

```python
# посмотрим на расспределение колонки 'limit'
# чтобы понять, какое количество пользователей
# имеют максимальный лимит

cust_ac.limit.hist(bins=30)
```

```python
# в данном случае колонка несет мало
# информации потому, что большенство
# имеет максимальный лимит и всего в этой
# колонке около трех уникальных значения

# расспечатаем их
cust_ac.limit.value_counts()
```

```python
# уникальных значений больше, но 
# отличаються они 1-3 у.е.

# отформатируем такие записи
cust_ac['limit'] = cust_ac.limit.apply(lambda x: round(x, -3))

# проверим новое распределение
# колонки 'limit'
cust_ac.limit.value_counts()
```

```python
# посмотрим на транзакции пользователей,
# которые младше 18 лет

under_18_spend = spend.query('Customer in @under_18.customer')

display(under_18_spend.head(10), under_18_spend.shape)
```

```python
# посмотрим на распределение стоимости
# транзакций от пользователей старше и
# младше 18 лет

over_18_spend = spend.query('Customer not in @under_18.customer')

fig, ax = plt.subplots()

bins = 10
alpha = 0.5

plt.hist(under_18_spend.Amount, bins=bins,
        alpha=alpha, density=True, label='under 18')
plt.hist(over_18_spend.Amount, bins=bins,
        alpha=alpha, density=True, label='over 18', )

plt.title('Стоимость транзакции для пользователей разного возраста')
plt.xlabel('стоимость')
plt.ylabel('частота')

plt.legend()

plt.show()
```

Можно предположить, что эти пользователи более поляризованы в своих затратах. Но не стоит забывать, что всего транзакций 70 и при увеличении количества, график выровняется по отношению пользоваетелей старше 18 лет.

На основе типа кредитной карточки, место работы и стоимости транзакций как у других пользователей, я делаю вывод, что возраст младше 18 лет это ошиюка в данных, которую нужно исправить. Заменим возраст у таких пользователей на среднее значение с математическим округлением.

```python
cust_ac.loc[cust_ac['age'] < 18, 'age'] = cust_ac.query('age > 18').age.mean().round().astype('int')
```

```python
# проверим на ошибки

cust_ac.query('age < 18')
```

### Repaymant

```python
# изменим названия колонок
repayment.columns = [i.lower() for i in repayment.columns]

# удалим колонки с названием 'sl no:' и 'unnamed: 4'
repayment.drop(['sl no:', 'unnamed: 4'], axis=1, inplace=True)

# проверим результат
repayment.head()
```

```python
# изменим тип колонки 'month' на
# 'datetime' и переименуем колонку в 'date'

repayment['month'] = pd.to_datetime(repayment['month'])
repayment.rename(columns={'month': 'date'}, inplace=True)
```

```python
# проверим результат

repayment.info()
```

```python
# посмотрим на пропуски в данных

repayment[repayment.isna().any(axis=1)]
```

```python
# пустые строки заменить не получится,
# поэтому удаляем их

repayment = repayment[~repayment.isna().any(axis=1)]
```

```python
# проверка

repayment.isna().sum()
```

```python
# посмотрим на дубликаты

repayment[repayment.duplicated()]
```

Получается строки дубликаты были строками с пустыми значениями. Удалив вторые, мы решали проблему с первыми.
