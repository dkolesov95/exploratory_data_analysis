# 1. Обзор данных

```python
# создадим функцию, которая будет 
# печатать базовую информацию
# о каждой таблице

path = '/Users/dmytro/Downloads/datasets/credit_cards_eda/'

def first_view() -> list:
    """
    Считывает файлы из указанной директории и отображает информацию о каждом файле.
    Функция выводит первые 5 строк таблицы, информацию о типах данных и статистические данные.
    Также функция выводит количество пропущенных значений и дубликатов в таблице.

    :return: Список pandas.DataFrame, содержащий данные из всех файлов директории.
    :rtype: list
    """
    files = [i for i in os.listdir(path) if not i.startswith('.')]
    dfs = []
    for file in files:
        df = pd.read_csv(path+file)
        dfs.append(df)
        
        print(file, '\n')
        display(
            df.head(),
            df.info(),
            df.describe(include='all')
        )
        print(
            f'Количество пропущеных значений:\n{df.isna().sum()}\n\n'
            f'Количество дубликатов - {df.duplicated().sum()}\n\n'
        )
        
    return dfs

files = first_view()
```

```python
# посмотрим на средний возраст в разрезе
# города, типа карты и типа компании

def barh_plot(main_col: str, label: str, add_col: str='age') -> None:
    """
    Создает горизонтальную столбчатую диаграмму, отображающую количество элементов каждой категории
    из столбца main_col, а также среднее значение столбца add_col для каждой категории.

    :param main_col: Имя столбца, содержащего категории для построения диаграммы.
    :type main_col: str
    :param label: Метка для значения, которое будет отображаться рядом с каждой категорией.
    :type label: str
    :param add_col: Имя столбца, для которого будет вычисляться среднее значение.
        По умолчанию 'age'.
    :type add_col: str
    :return: None
    :rtype: None
    """
    mean_age = cust_ac.groupby(main_col)[add_col].mean().to_dict()

    y_info = cust_ac[main_col].value_counts().sort_values()
    y_pos = np.arange(len(y_info.index))

    fig, ax = plt.subplots()

    barh = ax.barh(y_pos, y_info.values, align='center')
    ax.set_yticks(y_pos, y_info.index)

    for i, (k, v) in enumerate(y_info.to_dict().items()):
        value = f'{round(mean_age[k])} {label}'
        ax.text(v+0.5, i, value, va='center')

    ax.set_xlim(right=y_info.values.max()+5)
```

```python
barh_plot('city', 'св')

plt.title('Количество пользователей по городам')
plt.xlabel('пользователи')
plt.ylabel('город')

plt.legend(title='св - средний возраст', loc='lower right')

plt.show();

display(cust_ac.city.value_counts())
```

### Spend

Объединить таблицу с данными о транзакциях и информацией о пользователях, чтобы проанализировать следующее:
* расходы владельцев разных типов карт, жителей разных городов и разного возраста,
* что чаще всего покупают владельцы карт разного типа,
* в какие даты наблюдаются наибольшие расходы;

```python
cust_spend = pd.merge(left=cust_ac, right=spend, how='left', on='customer')

cust_spend.head()
```

```python
def amount_plot(key: str) -> None:
    """
    Отображает расходы клиентов по заданному ключу.

    :param key: Ключ для группировки расходов клиентов.
    :type key: str
    :return: Нет возвращаемого значения.
    :rtype: None
    """
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    grb_value = cust_spend.groupby(key) \
                            .amount \
                            .sum() \
                            .sort_values()

    grb_value.plot(kind='barh', ax=axs[0])
    
    cust_spend.pivot_table(
        values='amount',
        index='customer',
        columns=key
    )[grb_value.index[::-1]].plot(kind='box', ax=axs[1])

    return fig, axs

```

```python
fig, axs = amount_plot('product')

fig.suptitle('Сумма транзакций по типу карты')

axs[0].set(
    xlabel='сумма',
    ylabel='карта'
)

axs[1].set(
    xlabel='карта',
    ylabel='сумма'
)

plt.show();
```

```python
cust_spend.pivot_table(
        values='amount',
        index='customer',
        columns='product'
    )[['Gold', 'Platimum', 'Silver']].describe()
```

1. Денежный оборот у карт Gold и Platinum примерно одинаковый, и значительно превышает оборот у карт Silver, который в два раза меньше.
2. Медиана трат для всех трех типов карт примерно одинакова, что может свидетельствовать о схожих потребностях и поведении пользователей.
3. Карты Gold и Silver имеют большое количество аномально больших и маленьких транзакций, тогда как карты Platinum имеют меньше выбросов, что может указывать на более консервативное поведение пользователей.
4. Распределение трат для карт Silver характеризуется длинными хвостами, тогда как распределение для карт Gold и Platinum более сгруппированное и менее разнообразное.

```python
fig, axs = amount_plot('city')

fig.suptitle('Сумма транзакций по городам')

axs[0].set(
    xlabel='сумма',
    ylabel='город'
)

axs[1].set(
    xlabel='город',
    ylabel='сумма'
)

axs[1].tick_params(axis='x', rotation=45)

plt.show();
```

```python
cust_spend['age_bins'] = pd.cut(cust_spend.age, 6).astype('str')

fig, axs = amount_plot('age_bins')

fig.suptitle('Сумма транзакций по возрасту')

axs[0].set(
    xlabel='сумма',
    ylabel='возраст'
)

axs[1].set(
    xlabel='возраст',
    ylabel='сумма'
)

axs[1].tick_params(axis='x', rotation=45)

plt.show();
```

```python
(
    pd.crosstab(
        index=cust_spend.age_bins, 
        columns=cust_spend['product'], 
        aggfunc='count', 
        values=lambda x: 1, 
        normalize='index'
    ).plot(kind='barh')
)

plt.title('Распределение владельцев банковских карт по возрасту')
plt.xlabel('частота')
plt.ylabel('возраст')

plt.show();
```

1. Молодые люди в возрасте от 18 до 29 лет являются самыми активными пользователями банковских карт. Среди людей в возрасте от 30 до 59 лет обороты примерно одинаковы, в то время как люди в возрасте от 60 до 79 лет пользуются картами меньше, чем остальные группы.
2. Стоимость транзакций распределена примерно одинаково.
3. Возрастная группа от 18 до 29 лет в основном пользуется картами Gold, в то время как группы от 30 до 39 лет и от 70 до 79 лет предпочитают карты Silver и Gold. Люди в возрасте от 40 до 69 лет предпочитают карты Platinum и Gold, при этом группа в возрасте от 50 до 59 лет пользуется как картами Gold, так и Platinum.

```python
spend_by_type = pd.crosstab(
    index=cust_spend.type, 
    columns=cust_spend['product'], 
    aggfunc='count', 
    values=lambda x: 1, 
    normalize='index'
)

sns.heatmap(spend_by_type, cmap='crest', annot=True)

plt.title('Тепловая карта по категориям покупок'),
plt.xlabel('карта')
plt.ylabel('тип покупки')

plt.show();
```

1. Владельцы карт типа Gold предпочитают тратить больше на авиабилеты, автобусные билеты, фотоаппараты, petro, аренду жилья, сандалии и ж/д билеты.
2. Владельцы карт типа Platinum предпочитают тратить больше на автомобили, мотоциклы, одежду, ювелирные изделия и билеты в кино.
3. У владельцев карт типа Silver можно выделить предпочтения по таким категориям, как сандалии, ж/д билеты, билеты в кино и фотоаппараты.

```python
cust_spend['date_month'] = pd.to_datetime(cust_spend.date.dt.strftime('%m-%Y'))

(
    cust_spend
        .pivot_table(values='amount', index='date_month', columns='product')
        .reset_index()
        .sort_values('date_month')
        .groupby('date_month')
        .sum()
        .plot(figsize=(14,6))
)

plt.title('Динамика оборота по месяцам')
plt.xlabel('дата')
plt.ylabel('оборот')

plt.show();
```

1. Можно увидеть небольшой тренд на снижение оборота за период с 2004-01 по 2007-01.

В данных представлена информация о пользователях, включая их возраст, место жительства, тип кредитной карты, лимит, тип компании в которой работают и информация о транзакциях.

Большинство пользователей проживает в четырех городах (COCHIN, BANGALORE, CALCUTTA, BOMBAY).

Распределение пользователей по типам карт почти равномерное. Наибольшее число пользователей (38) приходится на средний сегмент карт (Gold), в то время как низкий сегмент (Silver) и премиальный сегмент (Platinum) имеют 30 и 32 пользователя соответственно. Владельцы карт Silver и Gold имеют одинаковый средний возраст. В то же время, владельцы премиальных карт Platinum в среднем моложе на 5-6 лет.

Люди более возрастной категории имеют самые высокие лимиты (от 100к до 500к). Средний возраст этой группы составляет 50 лет. В то же время, люди младшей возрастной категории обладают лимитами в размере 10к. Средний возраст этой группы составляет 40 лет.

Средний кредитный лимит у владельцев карт Gold является наибольшим по сравнению с Platinum и Silver, у которых лимит примерно одинаковы.

Денежный оборот у карт Gold и Platinum примерно одинаковый, и значительно превышает оборот у карт Silver, который в два раза меньше.
Медиана трат для всех трех типов карт примерно одинакова, что может свидетельствовать о схожих потребностях и поведении пользователей.
Карты Gold и Silver имеют большое количество аномально больших и маленьких транзакций, тогда как карты Platinum имеют меньше выбросов, что может указывать на более консервативное поведение пользователей.

Молодые люди в возрасте от 18 до 29 лет являются самыми активными пользователями банковских карт. Среди людей в возрасте от 30 до 59 лет обороты примерно одинаковы, в то время как люди в возрасте от 60 до 79 лет пользуются картами меньше, чем остальные группы.
Стоимость транзакций распределена примерно одинаково.
Возрастная группа от 18 до 29 лет в основном пользуется картами Gold, в то время как группы от 30 до 39 лет и от 70 до 79 лет предпочитают карты Silver и Gold. Люди в возрасте от 40 до 69 лет предпочитают карты Platinum и Gold, при этом группа в возрасте от 50 до 59 лет пользуется как картами Gold, так и Platinum.

Владельцы карт типа Gold предпочитают тратить больше на авиабилеты, автобусные билеты, фотоаппараты, petro, аренду жилья, сандалии и ж/д билеты.
Владельцы карт типа Platinum предпочитают тратить больше на автомобили, мотоциклы, одежду, ювелирные изделия и билеты в кино.
У владельцев карт типа Silver можно выделить предпочтения по таким категориям, как сандалии, ж/д билеты, билеты в кино и фотоаппараты.
